```python
import pandas as pd
import numpy as np
```

```python
dataset = {
    'Name': ['Keshav', 'Anuj', 'Prathamesh', 'Adinath', 'Aditya', 'Harsh', 'Shreyas', 'Tejas'],
    'Math': [122, 78, 86, 84, 10, 68, 78, np.nan],
    'Science': [np.nan, 45, 87, 65, 85, 255, 10, 87],
    'CGPA': [120, np.nan, 8, 8.87, 7.65, 5.59, 8.45, 9.84]
}

df = pd.DataFrame(dataset)
df
```

```python
df.isnull().sum()
```

```python
df.loc[df['Math'] > 100, 'Math'] = 100
df.loc[df['Science'] > 100, 'Science'] = 100
df.loc[df['CGPA'] > 10, 'CGPA'] = 10

df
```

```python
df['Math'] = df['Math'].fillna(df['Math'].mean())
df['Science'] = df['Science'].fillna(df['Science'].mean())
df['CGPA'] = df['CGPA'].fillna(df['CGPA'].mean())

df
```

```python
Q1 = df['Math'].quantile(0.25)
Q3 = df['Math'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print('Lower: ', lower)
print('Upper: ', upper)
```

```python
df = df[(df['Math'] >= lower) & (df['Math'] < upper)]
df
```

```python
Q1 = df['Science'].quantile(0.25)
Q3 = df['Science'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print('Lower: ', lower)
print('Upper: ', upper)

df = df[(df['Science'] >= lower) & (df['Science'] < upper)]
df
```

```python
Q1 = df['CGPA'].quantile(0.25)
Q3 = df['CGPA'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print('Lower: ', lower)
print('Upper: ', upper)

df = df[(df['CGPA'] >= lower) & (df['CGPA'] < upper)]
df
```

```python
df['Normalize_Math'] = (df['Math'] - df['Math'].min())/(df['Math'].max() - df['Math'].min())
df['Normalize_Science'] = (df['Science'] - df['Science'].min())/(df['Science'].max() - df['Science'].min())
df['Normalize_CGPA'] = (df['CGPA'] - df['CGPA'].min())/(df['CGPA'].max() - df['CGPA'].min())
```
